```python
import pandas as pd
```

```python
##1
students = pd.read_csv("attendance.csv")
students = students.rename(columns = {'Unnamed: 0' : 'name'})
students = students.melt(id_vars=['name'])
students = students.rename(columns = {'variable' : 'date', 'value' : 'attendance'})
students = students.set_index('date')
students.head()
```

```python
students = students.melt(id_vars=['name'], var_name='attendance', value_name= 'value')
students = students.drop(columns = 'attendance')
students = students.rename(columns = {'value' : 'attendance'})
```

```python
students = students.set_index('name')
students = students.attendance.replace(["P", "A", "T", "H"], [1, 0, .9, .5])
students = pd.DataFrame(students)
students.head()
```

```python
students = students.reset_index()
perc_attend = students.pivot_table(index=['name'], values = 'attendance')
perc_attend
```

```python
#3a
recipe = pd.DataFrame(pd.read_csv("cake_recipes.csv"))
recipe
```

```python
#3b
split = recipe["recipe:position"].str.split(":", n = 1, expand = True)
split
```

```python
recipe = pd.concat([split, recipe], axis=1).drop(columns='recipe:position')
recipe.head()
```

```python
recipe = recipe.rename(columns = {0 : 'recipe', 1 : 'rack_location'})
recipe.head()
```

```python
recipe = recipe.melt(id_vars=['recipe','rack_location'], var_name='temp', value_name= 'score')
recipe.head()
```

```python
#3c
best_recipe = recipe[recipe.score == recipe.score.max()]
best_recipe
```

```python
#3d
temp_avgs = recipe.pivot_table(index=['temp'], values = 'score')
best_temp = temp_avgs[temp_avgs.score == temp_avgs.score.max()]
best_temp
```

```python
#3e
avgs = recipe.pivot_table(index=['recipe', 'rack_location', 'temp'], values = 'score')
best_combo = avgs[avgs.score == avgs.score.max()]
best_combo
```
